# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("summer.csv")
#df.isnull().sum()
```

### 1. In how many cities Summer Olympics is held so far?

```python
print("Total Number of Cities is :",len(df['City'].unique()))
print(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
f = df[(df['Medal'] == 'Gold') ]
data=[]
for brand in f['Sport'].unique():
    data.append([brand , len(f[f['Sport']  == brand])])
#print(data)    
print(pd.DataFrame(data,columns = ['Sports','Medals']).sort_values(by='Medals', ascending=False).head())
pd.DataFrame(data,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sport', y = 'Medals', kind = 'bar', figsize = (10,10))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for brand in df['Sport'].unique():
    data.append([brand , len(df[df['Sport']  == brand])])
#print(data)
print(pd.DataFrame(data,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head())

pd.DataFrame(data,columns = ['sport','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'sport', y = 'Medals', kind = 'bar', figsize = (10,10))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for brand in df['Athlete'].unique():
    data.append([brand , len(df[df['Athlete']  == brand])])
#print(data)    
print(pd.DataFrame(data,columns = ['Player','Medals']).sort_values(by='Medals', ascending=False).head())
pd.DataFrame(data,columns = ['Player','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Player', y = 'Medals', kind = 'bar', figsize = (10,10))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
d=df[(df["Country"]=="IND") & (df["Medal"]=="Gold")]
year=d["Year"]
print("India won first gold medal in ",year.min())
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for brand in df['Event'].unique():
    data.append([brand , len(df[df['Event']  == brand])])
#print(data)    
print(pd.DataFrame(data,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head())
pd.DataFrame(data,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players', kind = 'bar', figsize = (10,10))
```
